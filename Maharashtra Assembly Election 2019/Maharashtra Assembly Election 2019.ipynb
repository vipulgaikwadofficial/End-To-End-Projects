# Maharashtra Assembly Election 2019 


## Overview

### Background

- Maharashtra Assembly Election 2019 were one of the most important elections in INDIA. 
- There were around 5 Crore voters, more than 100 Parties and 3000+ Candidates contesting election.
- On news and other platforms other aspects of election results are not highlighted which I will be covering here.
- Understanding election data can make huge change and candidate can focus on their next election campaigns.

### Objective 

- Clean and transform data for visualisation.
- Explore the dataset to understand the most common attributes of Maharashtra Assembly Election 2019.

### The Data 

- This data has been downloaded from Election Commission Of India Website.
- This is the most raw form of data available for Indian Elections.

     - State Name :-  Contains State or the Union Territory name.
     - AC No.     :-  Contains Assembly constituency number.
     - AC Name    :-  Contains Assembly Constituency Name.
     - Candidate Name :- Contains Contesting Candidates Name.
     - Sex        :-  Gender of Contesting Candidate.
     - Age        :-  Age of Contesting candidate.
     - Category   :-  Contains category of candidate.
     - Party      :-  Name of Party.
     - Symbol     :-  Party symbol
     - General    :-  Contains votes casted using EVM Machine
     - Postal     :-  Contains total votes casted through post
     - Total      :-  Sum of General and Postal votes casted.
     - % Votes Polled :- Percentage of votes polled Total/Total Elecotrs.
     - Total Electors :- Total Voters/Electors as per in Election Commissions Voters Database who were eligible to cast vote.

### Import Libraries

```python
#import libraries

import pandas as pd
```

```python
# read csv file

data = pd.read_csv('election.csv', skipinitialspace = True)
data
```

```python
# Convert to a dataframe

election_data = pd.DataFrame(data)
election_data
```

```python
# view info

election_data.info()
```

```python
# remove white space from column names

election_data.columns = election_data.columns.str.strip()
election_data
```

```python
# Rename columns

election_data.rename(columns = {'STATE/UT NAME':'state'},inplace =True)
election_data.rename(columns = {'AC NO.':'assembly_no'},inplace =True)
election_data.rename(columns = {'AC NAME':'assembly'},inplace =True)
election_data.rename(columns = {'CANDIDATE NAME':'candidate_name'},inplace = True)
election_data.rename(columns = {'SEX':'sex'},inplace=True)
election_data.rename(columns = {'AGE':'age'},inplace=True)
election_data.rename(columns = {'CATEGORY':'category'},inplace =True)
election_data.rename(columns = {'PARTY':'party'},inplace =True)
election_data.rename(columns = {'SYMBOL':'symbol'},inplace =True)
election_data.rename(columns = {'GENERAL':'general'},inplace=True)
election_data.rename(columns = {'POSTAL':'postal'},inplace =True)
election_data.rename(columns = {'TOTAL':'total'},inplace =True)
election_data.rename(columns = {'% VOTES POLLED':'%_votes_polled'},inplace =True)
election_data.rename(columns = {'TOTAL ELECTORS':'total_electors'},inplace =True)
election_data
```

#### There are blank rows in this data. Drop all the blank rows

```python
# Remove whitespace's from all columns

election_data['state'] = election_data['state'].str.strip()
election_data
```

```python
# Remove Null Rows

turnout = election_data[election_data['state']=='TURNOUT']
turnout
```

```python
# Drop rows containing TURNOUT 

election_data.drop(election_data[election_data['state']=='TURNOUT'].index,inplace = True)
election_data
```

```python
# drop last row

election_data.drop(election_data.tail(1).index, inplace = True)
election_data
```

#### There are some rows of NOTA where there are no values in cell. Which is totally fine.

```python
election_data.info()
```

### Next in the candidate_name column there are numbers according to their rank. Ex. 1 ADV. K. C. PADAVI, Remove those numbers

```python
# copy dataframe

df1 = election_data.copy()
df1
```

```python
# remove numbers from data

df1['candidate_name'] = df1['candidate_name'].str.replace('\d+','')
df1
```

```python
# Export to csv

df1.to_csv('final_election_data.csv')
```
